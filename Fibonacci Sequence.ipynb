```python
MAX_TERMS = 10000


def fibonacci_to_value(limit):
    if limit < 0:
        return []
    if limit == 0:
        return [0]

    sequence = [0, 1]
    while True:
        next_value = sequence[-1] + sequence[-2]
        if next_value > limit:
            break
        sequence.append(next_value)

    return sequence


def fibonacci_n_terms(n):
    if n <= 0:
        return []
    if n == 1:
        return [0]

    sequence = [0, 1]
    while len(sequence) < n:
        sequence.append(sequence[-1] + sequence[-2])

    return sequence


def format_sequence(sequence):
    return ", ".join(str(value) for value in sequence)


print("Up to 21:", format_sequence(fibonacci_to_value(21)))
print("First 8 terms:", format_sequence(fibonacci_n_terms(8)))


print("Choose mode:")
print("1) Generate Fibonacci sequence up to a maximum value")
print("2) Generate first N Fibonacci terms")
mode = input("Enter 1 or 2: ").strip()

if mode == "1":
    raw = input("Enter maximum value (0 or more): ").strip()
    try:
        maximum_value = int(raw)
    except ValueError:
        print("Please enter a whole number.")
    else:
        if maximum_value < 0:
            print("Please enter 0 or more.")
        else:
            sequence = fibonacci_to_value(maximum_value)
            print(f"Fibonacci sequence up to {maximum_value}:")
            print(format_sequence(sequence))
elif mode == "2":
    raw = input(f"Enter number of terms (1-{MAX_TERMS}): ").strip()
    try:
        terms = int(raw)
    except ValueError:
        print("Please enter a whole number.")
    else:
        if terms <= 0:
            print("Please enter 1 or more.")
        elif terms > MAX_TERMS:
            print(f"Too large. Max allowed is {MAX_TERMS}.")
        else:
            sequence = fibonacci_n_terms(terms)
            print(f"First {terms} Fibonacci terms:")
            print(format_sequence(sequence))
else:
    print("Invalid option. Please run again and enter 1 or 2.")
```
